```python
import pandas as pd
import numpy as np
import folium 
```

```python
data = pd.read_csv('data/Parcels.csv')
```

```python
#Drop missing zip codes and sales price
data = data.replace({'ZIPNUM': '0'}, np.nan)
dropped_zip = data.dropna(subset=['ZIPNUM'])
dropped_zip = dropped_zip.dropna(subset=['TOTSALPRICE'])


#Looking at only residential data and sales after 2019
dropped_zip = dropped_zip.loc[dropped_zip['LAND_CLASS_DECODE'] == 'Residential Less Than 10 Acres']
dropped_zip = dropped_zip.loc[dropped_zip['DEED_DATE'] > '2019/01/01 00:00:00+00']
dropped_zip
```

```python
dropped_zip.astype({'TOTAL_VALUE_ASSD': 'float'}).dtypes
group = dropped_zip.groupby(["ZIPNUM"])
group = group["TOTSALPRICE"].mean()
group = round(group.reset_index(), 0)
group['ZIPNUM'] = group['ZIPNUM'].astype(str)
#group.to_csv('data/zip_by_sales_price_2019.csv')
data = pd.read_csv('data/zip_by_sales_price_2019.csv')
data['ZIPNUM'] = data['ZIPNUM'].astype(str)
data['TOTSALPRICE'] = round(data['TOTSALPRICE'], -3)
data
```

```python
map = folium.Map(location=[35.8032, -78.5661], default_zoom_start=15)
map.choropleth(geo_data="data/nc_zip_geo2.json",
             data=data, # my dataset
             columns=['ZIPNUM', 'TOTSALPRICE'],
             key_on='feature.properties.ZCTA5CE10', # this path contains zipcodes in str type, this zipcodes should match with our ZIP CODE column
             fill_color='BuPu', fill_opacity=.7, line_opacity=.2,
             legend_name='Mean Sales Price')
map
map.save('index.html')
```
